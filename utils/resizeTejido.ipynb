```
from PIL import Image


path = "tissues_train/"
directory_files = os.listdir(path)
multiple_images = [file for file in directory_files if file.endswith(('.jpg', '.png'))]
def resize():
    for item in directory_files:
        if os.path.isfile(path+item) and item.endswith(('.jpg', '.png')):
          im = Image.open(path+item)
          f, e = os.path.splitext(path+item)
          imResize = im.resize((1000,1000), Image.ANTIALIAS)
          imResize.save("/content/drive/My Drive/Datos/Gleason/resized_train/"+item, optimize=True, quality=90)

resize()
```
