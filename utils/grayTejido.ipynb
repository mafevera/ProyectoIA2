```
import cv2


from skimage.color import rgb2gray

path = "tissues_train/"
directory_files = os.listdir(path)
multiple_images = [file for file in directory_files if file.endswith(('.jpg'))]
def resize():
    for item in directory_files:
        if os.path.isfile(path+item) and item.endswith(('.jpg')):
          #comando para leer imagen
          im = cv2.imread(path+item, cv2.IMREAD_UNCHANGED)
          im_gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
          # mostrar una imagen
          #Guardar una imagen
          
          cv2.imwrite("/content/drive/My Drive/Datos/Gleason/gray_train/"+item,im_gray)

resize()
```

```
import json

path = "tissues_train/"
directory_files = os.listdir(path)
def resize():
    for item in directory_files:
        if os.path.isfile(path+item) and item.endswith(('.png')):
          #comando para leer imagen
          im = cv2.imread(path+item, cv2.IMREAD_UNCHANGED)
          
          # mostrar una imagen
          #Guardar una imagen
          
          cv2.imwrite("/content/drive/My Drive/Datos/Gleason/gray_train/"+item,im)
        elif os.path.isfile(path+item) and item.endswith(('.json')):
          with open(os.path.join(path + item),'r') as file:
            data = json.load(file)
          with open(os.path.join("/content/drive/My Drive/Datos/Gleason/gray_train/",item),'w') as file:
            json.dump(data, file, indent=4)


resize()
```
