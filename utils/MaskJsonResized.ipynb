```
import cv2, json
import numpy as np
from google.colab.patches import cv2_imshow

#!wget https://www.dropbox.com/s/ocuwkmgigzvsu6e/mask_ZT76_39_B_2_5.png?dl=0 -O mask_ZT76_39_B_2_5.png
#!wget https://www.dropbox.com/s/excbs4zq92kxf02/mask_ZT76_39_A_1_3.png?dl=0 -O mask_ZT76_39_A_1_3.png
#!wget https://www.dropbox.com/s/gf2wi9mz1dhrmos/mask_ZT76_39_A_1_12.png?dl=0 -O mask_ZT76_39_A_1_12.png


def contornos_por_color(imagen,color):
	img = cv2.imread(imagen)
	imghsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

	if color=="amarillo":
		lower = np.array([29,255,255])
		upper = np.array([30,255,255])

	if color=="azul":
		lower = np.array([110,255,255])
		upper = np.array([130,255,255])

	if color=="rojo":
		lower = np.array([0,255,255])
		upper = np.array([10,255,255])
    
	if color=="verde":
		lower = np.array([55,255,255])
		upper = np.array([65,255,255])

	mask = cv2.inRange(imghsv, lower, upper)
	contours, hierarchy = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
	poligono,poligonos=[],[]

	for k in range(len(contours)):
	  for i in range(len(contours[k])):
	    poligono.append([contours[k][i][0][0],contours[k][i][0][1]])
	  poligonos.append(poligono[1::2])
	  poligono = []

	
	for i in range(len(poligonos)):
	  shapes.append(
	    {
	      "label": color,
	      "line_color": None,
	      "fill_color": None,
	      "points": [[int(k[0]),int(k[1])] for k in poligonos[i]],
	      "shape_type": "polygon"
	    }
	  )

	return shapes
```

```
import os

path, dirs, files = next(os.walk("."))
for file in files:
    if file.endswith(".png"):
      shapes=[]
      
      shapes_ = contornos_por_color(file,"amarillo")
      shapes_ = contornos_por_color(file,"azul")
      shapes_ = contornos_por_color(file,"rojo")
      shapes_ = contornos_por_color(file,"verde")


      diccionario = {
        "version": "3.14.2",
        "flags": {},
        "shapes": shapes,
        "lineColor": [
          0,
          255,
          0,
          128
        ],
        "fillColor": [
          255,
          0,
          0,
          128
        ],
        "imagePath": file,
        "imageHeight": 3100,
        "imageWidth": 3100
      }

      with open(file[:-4]+'.json', 'w') as json_salida:
          json.dump(diccionario, json_salida, indent=4)
      print("OK - ",file)
```
